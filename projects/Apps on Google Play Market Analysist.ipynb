```python
## I notice that Size has different format. We know that 1 MB = 1024 KB. So we agree that we convert all units to MB
## I check that Size, Installs, Price have a specical character. It will impede their conversion to a numerical data type. So I will solve 
## them
special_characters = ["M", "$", "+", ","]
target_numerical_cols = ["Size", "Installs", "Price"]
for col in target_numerical_cols:
    for character in special_characters:
        if (character != "M") & (character != ","):
            character = "\\" + character
        if playstore_df[col].str.contains(character).any():
            playstore_df[col] = playstore_df[col].str.replace(character, "", regex = True)
pd.concat([playstore_df.head(), playstore_df.tail()], axis = 0)          
```

```python
## I check that Genres based on Category but they don't match each other so I will solve it
def custom_format(text):
    return " ".join([word.title() for word in text.split()])
```

```python
playstore_df["Category"] = playstore_df["Category"].astype("string") ## That's why object type gets on my nerves when working with data
playstore_df["Category"] = playstore_df["Category"].str.replace("_", " ", regex = True)
playstore_df["Category"] = playstore_df["Category"].str.replace("AND", "&", regex = False)
playstore_df["Category"] = playstore_df["Category"].apply(custom_format)
playstore_df[["Category", "Genres"]].head()
```

### **`5) Check Null Value`**

```python
def data_info():
    tmp = pd.DataFrame(index = playstore_df.columns)
    tmp["Data Type"] = playstore_df.dtypes
    tmp["Number Of Not Null Value"] = playstore_df.count()
    tmp["Number Of Null Value"] = playstore_df.isnull().sum()
    # tmp["% Of Null Value"] = playstore_df.isnull().mean() * 100
    tmp["% Of Null Value"] = round(playstore_df.isnull().mean() * 100, 2)
    tmp["Number Of Unique Value"] = playstore_df.nunique()
    return tmp
```

### **`7) Problem Statement`**

```python
## Question 1: How much is maximum potential users for each target user that we can reach ? How much is maximum capacity that each target's device have
## (Assume that for each install we have 1 unique user)
## Standard unit potential user I will choose million
standard_unit = 1000000
target_user_df = playstore_df.groupby("Content Rating", as_index = False).agg({"Size": "max", "Installs": "max"})
target_user_df = target_user_df.rename(columns = {"Size": "Maximum Capacity User's Device Have", "Content Rating": "Target User"})
target_user_df = target_user_df.rename(columns = {"Installs": "Maximum Potential Users", "Size": "Maximum Capacity User's Device Have", "Content Rating": "Target User"})
target_user_df = target_user_df.sort_values(by = "Maximum Potential Users", ascending = False)
target_user_df["Maximum Potential Users"] = target_user_df["Maximum Potential Users"] / standard_unit
# target_user_df["Maximum Potential Users"] = target_user_df["Maximum Potential Users"].apply(lambda x: "{:,}".format(x))
target_user_df
```

```python
## Visualize 
plt.figure(figsize=(20,6))
ax = sns.barplot(x = target_user_df["Maximum Potential Users"], y = target_user_df["Target User"])
for container in ax.containers:
    ax.bar_label(container, fmt = "%.2fM")
plt.title("Maximum Potential Users for each Target User (In Millions)")
plt.gca().set_xticks([])
plt.show()
```

```python
## Visualize
plt.figure(figsize=(20,6))
ax = sns.barplot(x = target_user_df["Maximum Capacity User's Device Have"], y = target_user_df["Target User"])
for container in ax.containers:
    ax.bar_label(container, fmt = "%.1fMB")
plt.title("Maximum Capacity User's Device have")
plt.gca().set_xticks([])
plt.show()
```

**Key findings:** Through two above charts, I think I won't `invest` in a game that has a target user `Adults only 18+` or `Unrated`. Why? Firstly,
their maximum potential users is `equal` or `below 1 millions`. It's `too small` to be considered as potential segement as others.
Next, their capacity is extremely poor compared with others. Please remember, capacity is one of the `essential factors` that
helps us to `improve` `user experience`. In conclusion, ideal potential segments are `Everyone 10+`, `Teen` and `Mature 17+` (Why I don't
choose Everyone because I think `focusing on caring specific target user` is more `efficient` than `trying to target everyone`.

```python
## Question 3: How user consider "Affordable App"
tmp_df = playstore_df[playstore_df["Type"] == "Paid"][["Installs", "Price"]].sort_values(by = "Installs", ascending = False)
pd.concat([tmp_df.head(), tmp_df.tail()], axis = 0)
```

```python
def convert_number_unit(val):
    len_lst = len(val)
    if len_lst <= 3:
        return val
    result = ""    
    while(len(val) > 3):
        pos = len(val) - 3
        result += "." + val[pos:]
        val = val[:pos]
    return val + result    
```

```python
## After encountering overlapping axis tick label, I realize the problem is because it follows the number value. 
## So to solve it, we just need to convert it into non-numerical data
rating_and_install = playstore_df[["Installs", "Rating", "Reviews"]].copy()
rating_and_install = rating_and_install.sort_values(by = "Installs", ascending = True)
rating_and_install["Installs"] = rating_and_install["Installs"].astype("string")
rating_and_install["Installs"] = rating_and_install["Installs"].apply(lambda x: convert_number_unit(x))
rating_and_install.info()
```

```python
plt.scatter(rating_and_install["Installs"], rating_and_install["Rating"])
plt.xticks(rotation = 90)
plt.xlabel("Installs")
plt.ylabel("Rating")
plt.title("Rating vs Installs")
plt.show()
```

**Key findings:** Even though some apps has `impressive installs`, they still have a `poor rating`. This may be true, 
 because some `factors` can `affect` the `Rating` such as `Size`, `Android Ver`, `etc`.
But we `still see` some popular apps have incredible ratings (`~4.5`)
